```python
# Currently installed: 
# ensembl103 ensembl107 ensembl109 ensembl111 ensembl78  ensembl80  ensembl82  ensembl87  ensembl97
# ensembl105 ensembl108 ensembl110 ensembl77  ensembl79  ensembl81  ensembl83  ensembl93
import pyensembl 
def gene_id_for_name(
        gene_name, 
        ensembl_genomes=[pyensembl.cached_release(release) for release in (108, 107, 105, 97, 93, 87, 83, 75)]):
    for genome in ensembl_genomes:
        try:
            genes = genome.genes_by_name(gene_name)
        except ValueError:
            genes = []
        if genes:
                return genes[0].id
```
